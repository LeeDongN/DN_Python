```python
gray_pixel = []
```

```python
img_color = cv.imread(r"C:\program1\image_raw\01-0001.png", cv.IMREAD_COLOR)
img_gray = cv.cvtColor(img_color, cv.COLOR_BGR2GRAY)

height, width = img_gray.shape
img_gray_pixel1 = np.zeros((height, width), np.uint8)
```
